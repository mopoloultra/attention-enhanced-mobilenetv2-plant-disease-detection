```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
original_dataset_dir = 'data/PlantVillage_raw'
base_dir = 'data/PlantVillage'

classes = os.listdir(original_dataset_dir)
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')
```

```python
for cls in classes:
    img_dir = os.path.join(original_dataset_dir, cls)
    images = os.listdir(img_dir)
    
    train_imgs, temp_imgs = train_test_split(images, test_size=0.3, random_state=42)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

    # Make folders and copy
    for subset, subset_imgs in zip([train_dir, val_dir, test_dir], [train_imgs, val_imgs, test_imgs]):
        cls_dir = os.path.join(subset, cls)
        os.makedirs(cls_dir, exist_ok=True)
        for img in subset_imgs:
            src = os.path.join(img_dir, img)
            dst = os.path.join(cls_dir, img)
            shutil.copyfile(src, dst)
```
